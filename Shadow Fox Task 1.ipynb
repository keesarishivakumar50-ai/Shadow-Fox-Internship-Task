```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
sns.set(style='white')
```

```python
data={'Roll No':[25,39,52,40,51,54,34,76],
      'Name':['Shiva','Surya','Manoj','Mahender','Hussain','Rudra','Vijay','Akshath'],
      'Marks':[78,56,69,88,63,87,65,55]}
df=pd.DataFrame(data)
df=df.sort_values(by='Roll No')
df=df.reset_index(drop=True)
df
```

**Line Chart**

*   Trend Analysis Over Time
*   Monitoring Changes

```python
#SeaBorn
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x='Roll No',y='Marks')
plt.title('Seaborn Line Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.show()

#Matplotlib
plt.figure(figsize=(10,5))
plt.plot(df['Roll No'],df['Marks'])
plt.title('Matplotlib Line Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.show()
```

**Bar Chart**
*   Comparing quantities across categories
*   Visualizing counts or totals

```python
#Seaborn
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Roll No',y='Marks',palette='viridis')
plt.title('Seaborn Bar Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.show()

#Matplotlib
colors= ['red','blue','green','purple','orange','cyan','magenta','brown']
df['Roll No'] = df['Roll No'].astype(str)
plt.figure(figsize=(10,5))
plt.bar(df['Roll No'],df['Marks'],color=colors,width=0.8)
plt.title('Bar Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.show()
```

**Scatter Plot**

*   Correlation analysis
*   Identifying clusters or outliers


```python
#Seaborn
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='Roll No',y='Marks',hue='Marks',palette='viridis')
plt.title('Seaborn Scatter Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.grid(True)
plt.show()

#Matplotlib
plt.figure(figsize=(10,5))
plt.scatter(df['Roll No'],df['Marks'],color='red',marker='o')
plt.title('Matplotlib Scatter Plot')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```
